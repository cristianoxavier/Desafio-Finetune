```python
import matplotlib.pyplot as plt
from datetime import datetime
import requests
```

```python
data = requests.get('http://127.0.0.1:8080/v1/food/graph').json()
```

```python
x = ["Carboidrato", "Gordura", "Proteina"]
y = [data['carboidrato'], data['gordura'], data['proteina']]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

# Dados para cada subplot
ax1.bar(x,y)
ax2.pie(y, labels=x, autopct='%1.1f%%', shadow=True, startangle=90)

ax1.set(title="Gráfico de Barras", ylabel="Quantidade")
ax2.set(title="Gráfico de Pizza", xlabel="Distribuiçao")

plt.show()
```
